```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
# Bernoulli

p = 0.3
samples_bernoulli = np.random.binomial(n=1, p=p, size=1000)

sns.histplot(samples_bernoulli, discrete=True, stat='probability')
plt.title(f"Bernoulli(p={p})")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

```python
n, p = 10, 0.3
samples_binomial = np.random.binomial(n=n, p=p, size=1000)

sns.histplot(samples_binomial, discrete=True, stat='probability')
plt.title(f"Binomial(n={n}, p={p})")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

```python
lam = 5
samples_poisson = np.random.poisson(lam=lam, size=1000)

sns.histplot(samples_poisson, discrete=True, stat='probability')
plt.title(f"Poisson(Î»={lam})")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()
```

```python
mu, sigma = 0, 1
samples_normal = np.random.normal(loc=mu, scale=sigma, size=1000)

sns.histplot(samples_normal, stat='density', kde=True, color='blue')
plt.title(f"Normal(Î¼={mu}, Ï={sigma})")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from scipy.stats import uniform

# Parameters: a (min), b (max)
a, b = 0, 1

# Generate data
data = uniform.rvs(loc=a, scale=b-a, size=1000)

# Plot the distribution
plt.hist(data, bins=30, alpha=0.7, density=True)
plt.title("Uniform Distribution (0, 1)")
plt.show()
```

#### Normal approximation to binomial

```python
def plot_binomial(n, p = 0.3):
    samples_binomial = np.random.binomial(n=n, p=p, size=1000)
    sns.histplot(samples_binomial, discrete=True, stat='probability')
    plt.title(f"Binomial(n={n}, p={p})")
    plt.xlabel("Number of Successes")
    plt.ylabel("Probability")
    plt.show()

for n in [10, 50, 200, 1000]:
    plot_binomial(n)
```

Some other distributions ...

```python
from scipy.stats import beta

# Parameters: alpha, beta
alpha, beta_param = 2, 5

# Generate data
data = beta.rvs(alpha, beta_param, size=1000)

# Plot the distribution
plt.hist(data, bins=30, alpha=0.7, density=True)
plt.title("Beta Distribution (alpha=2, beta=5)")
plt.show()
```

```python
from scipy.stats import t

# Parameters: degrees of freedom
df = 10

# Generate data
data = t.rvs(df, size=1000)

# Plot the distribution
plt.hist(data, bins=30, alpha=0.7, density=True)
plt.title(f"Student's t-Distribution (df={df})")
plt.show()
```
