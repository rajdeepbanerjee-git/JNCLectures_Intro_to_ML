```python
import numpy as np
import keras
import matplotlib.pyplot as plt
```

[Fashion MNIST dataset](https://keras.io/api/datasets/fashion_mnist/)

| Label | Description |
|:-----:|:-----------:|
|   0   | T-shirt/top |
|   1   | Trouser     |
|   2   | Pullover    |
|   3   | Dress       |
|   4   | Coat        |
|   5   | Sandal      |
|   6   | Shirt       |
|   7   | Sneaker     |
|   8   | Bag         |
|   9   | Ankle boot  |

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
x_train, x_test = x_train / 255., x_test / 255.
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# Create a figure and axes objects
num_examples = 5
fig, axs = plt.subplots(1, num_examples, figsize=(12, 4))

# Plot examples
for i in range(num_examples):
    axs[i].imshow(x_train[i], cmap='gray')
    axs[i].set_title('Label: {}'.format(y_train[i]))
    axs[i].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
nd = np.product(x_train.shape[1:])
```

```python
X_train, X_test = x_train.reshape([x_train.shape[0], nd]), x_test.reshape([x_test.shape[0], nd])
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
yhat_test = lr.predict(X_test)
```

```python
classification_report(y_test, yhat_test, output_dict=True)
```
