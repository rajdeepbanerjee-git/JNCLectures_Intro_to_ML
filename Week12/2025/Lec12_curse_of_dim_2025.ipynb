```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_features(N, p):
  features = np.ones(N)
  for dim in range(p):
    features = np.c_[features, np.random.rand(N)]

  return features
```

```python
def get_dist_ratio(A, N):

  dist_mat = []
  for i in range(N):
    dist_mat.extend(np.linalg.norm(np.delete(A, i, 0), axis = 1))

  dist_ratio = (max(dist_mat) - min(dist_mat))/min(dist_mat)

  return dist_ratio
```

```python
dist_ratio_ = []
n_samples = 1000
for dim in range(1, 201, 5):
  A = get_features(N = n_samples, p = dim)
  dist_ratio_.append(get_dist_ratio(A, N = n_samples))
```

```python
plt.plot(range(1, 201, 5), dist_ratio_)
```
