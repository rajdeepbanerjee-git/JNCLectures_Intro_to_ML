# Feature importance
In this exercise you will calculate the 'feature importance' (relative/absolute) for each of the features in a decision tree.

## Import the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

## Load data

```python
data = pd.read_csv('./fruit_quality.csv')
```

```python
X = data[data.columns[1:-1]].to_numpy()
y = data.Quality.to_numpy()
feature_names = data.columns[1:-1]
```

## Split data into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

## Train a decision tree

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
dtc = DecisionTreeClassifier() #(max_depth=3, min_samples_leaf=200)
```

```python
dtc.fit(X_train, y_train)
```

```python
plt.figure(figsize=(20,20),dpi=300)
plot_tree(dtc, feature_names = feature_names, node_ids=True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
```

```python
yhat_test = dtc.predict(X_test)
```

```python
accuracy_score(y_test, yhat_test)
```

```python
fig, ax = plt.subplots()

ax.bar(feature_names, dtc.feature_importances_)
ax.tick_params(axis='x', labelrotation=45)
ax.set_ylabel('Feature Importance')
fig.show()
```
